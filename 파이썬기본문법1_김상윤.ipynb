```python
# 사용자로부터 선박 이름, 길이, 폭, 흘수를 입력받습니다.
new_ship_name = input("새 선박 이름을 입력하세요: ")
new_length = float(input("새 길이 (L)를 입력하세요 (m): "))
new_breadth = float(input("새 폭 (B)을 입력하세요 (m): "))
new_draft = float(input("새 흘수 (T)를 입력하세요 (m): "))

# 방형비(Block Coefficient) Cb를 0.8로 가정합니다.
new_Cb = 0.8

# 배수량(Displacement)을 계산합니다.
# 공식: Displacement = L * B * T * Cb * 1.025 (해수의 밀도 1.025 t/m^3)
new_displacement = new_length * new_breadth * new_draft * new_Cb * 1.025

# 계산된 배수량을 출력합니다.
print(f"\n--- 새 선박 정보 ---")
print(f"선박 이름: {new_ship_name}")
print(f"길이 (L): {new_length} m")
print(f"폭 (B): {new_breadth} m")
print(f"흘수 (T): {new_draft} m")
print(f"계산된 배수량 (Displacement): {new_displacement:.2f} 톤")
```

## 1. 사용자로부터 정수를 2개 입력 받아 사칙 연산을 수행하기

```python
# 사용자로부터 첫 번째 정수 입력받기
a = int(input("첫 번째 정수를 입력하세요: "))

# 사용자로부터 두 번째 정수 입력받기
b = int(input("두 번째 정수를 입력하세요: "))

# 덧셈
sum_result = a + b
print(f"{a} + {b} = {sum_result}")

# 뺄셈
sub_result = a - b
print(f"{a} - {b} = {sub_result}")

# 곱셈
mul_result = a * b
print(f"{a} * {b} = {mul_result}")

# 나눗셈 (실수 결과)
# 0으로 나누는 경우를 방지하기 위해 조건문 추가
if b != 0:
    div_result = a / b
    print(f"{a} / {b} = {div_result}")

    # 나눗셈 (정수 몫)
    floor_div_result = a // b
    print(f"{a} // {b} = {floor_div_result}")

    # 나눗셈 (나머지)
    mod_result = a % b
    print(f"{a} % {b} = {mod_result}")
else:
    print("0으로는 나눌 수 없습니다.")
```

## 5. 거스름돈 계산기


역할: 너는 파이썬 튜터야

목표: 아래 문제를 파이썬으로 풀어줘.

문제: 거스름돈 계산기를 만들어줘. 사용자가 1만원 단위의 금액을 입력하면, 상품값을 제외한 거스름돈은 5천원, 1천원, 500원, 100원의 거스름돈을 받을 수 있도록 해줘.

입력: (예) 지불금액  : 20000 상품금액 : 15400

출력: (예)==거스름돈 ==
5,000원권 : 0장 1,000원권 : 4장 500원 동전 : 0개  100원 동전 : 3개 (각 권종당 줄바꿈으로 처리)

제약: 몫과 나머지 연산자 사용, 1000원 이상은 지폐 500원, 100원은 동전으로 처리, 사용자가 10,000원 단위로 입력할 수 있도록 제시.

요청: 코드에 대해 한 줄씩 설명하고 테스트 3개 실행


```python
# 사용자로부터 지불 금액을 입력받습니다. (예: 1만원 단위로 입력 유도)
payment_amount = int(input("지불 금액을 입력하세요 (예: 20000, 30000 등): "))
# 사용자로부터 상품 금액을 입력받습니다.
product_price = int(input("상품 금액을 입력하세요: "))

# 거스름돈을 계산합니다.
change_amount = payment_amount - product_price

print("\n== 거스름돈 ==")

# 5천원권 개수를 계산하고 남은 거스름돈을 업데이트합니다.
five_thousand_bills = change_amount // 5000
change_amount = change_amount % 5000

# 1천원권 개수를 계산하고 남은 거스름돈을 업데이트합니다.
one_thousand_bills = change_amount // 1000
change_amount = change_amount % 1000

# 5백원 동전 개수를 계산하고 남은 거스름돈을 업데이트합니다.
five_hundred_coins = change_amount // 500
change_amount = change_amount % 500

# 1백원 동전 개수를 계산하고 남은 거스름돈을 업데이트합니다.
one_hundred_coins = change_amount // 100
change_amount = change_amount % 100

# 각 권종별 거스름돈을 출력합니다.
print(f"5,000원권 : {five_thousand_bills}장")
print(f"1,000원권 : {one_thousand_bills}장")
print(f"500원 동전 : {five_hundred_coins}개")
print(f"100원 동전 : {one_hundred_coins}개")
```

## 6. 성적 등급 계산기 (A/B/C/D/F + 범위 체크)


### 문제 : 점수를 입력받아 학점 계산
- 입력: 0~100 점수(score)
- 출력: A(100-90), B(89-80), C(79-70), D(69-60), F(그 외)
- 추가: 범위 밖 입력이면 `"입력 오류"` 출력


```python
# TODO: 성적 등급 계산기 완성
score = int(input("score(0~100): "))

# 1) 범위 밖이면 오류
# 2) 범위 안이면 등급 계산
```

```python
# 사용자로부터 점수를 입력받습니다.
score = int(input("점수를 입력하세요 (0~100): "))

# 점수 범위 유효성 검사
if score < 0 or score > 100:
    print("입력 오류: 점수는 0에서 100 사이여야 합니다.")
else:
    # 학점 계산
    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    elif score >= 60:
        grade = 'D'
    else:
        grade = 'F'
    print(f"점수: {score}점, 학점: {grade}")
```

```python
# 점수 5개를 입력받아 리스트에 저장
scores_list = []
for i in range(5): # 5번 반복하여 점수를 입력받습니다.
    while True: # 유효한 점수(0~100)를 입력받을 때까지 반복합니다.
        try:
            score = int(input(f"{i+1}번째 점수를 입력하세요 (0~100): "))
            if 0 <= score <= 100: # 점수가 0에서 100 사이인지 확인합니다.
                scores_list.append(score) # 유효한 점수면 리스트에 추가합니다.
                break # 유효한 점수 입력 시 내부 루프를 탈출합니다.
            else:
                print("0점에서 100점 사이의 유효한 점수를 입력해주세요.") # 유효하지 않은 점수 범위 안내
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.") # 숫자가 아닌 입력 시 오류 안내

# for 문을 사용하여 합계, 평균, 최대값, 최소값, 90점 이상 개수 계산
if scores_list: # scores_list에 점수가 있는 경우에만 계산을 진행합니다.
    total_score = 0
    max_score = -1 # 최대값을 찾기 위한 초기값 (가장 낮은 점수보다 작게 설정)
    min_score = 101 # 최소값을 찾기 위한 초기값 (가장 높은 점수보다 크게 설정)
    count_above_90 = 0 # 90점 이상인 점수의 개수를 세기 위한 초기값

    for score in scores_list: # 리스트의 각 점수에 대해 반복합니다.
        total_score += score # 합계를 누적합니다.
        if score > max_score: # 현재 점수가 최대값보다 크면 최대값을 업데이트합니다.
            max_score = score
        if score < min_score: # 현재 점수가 최소값보다 작으면 최소값을 업데이트합니다.
            min_score = score
        if score >= 90: # 현재 점수가 90점 이상이면 개수를 1 증가시킵니다.
            count_above_90 += 1

    average_score = total_score / len(scores_list) # 평균을 계산합니다.

    print(f"\n점수 리스트: {scores_list}") # 입력된 점수 리스트를 출력합니다.
    print(f"점수 합계: {total_score}") # 계산된 합계를 출력합니다.
    print(f"점수 평균: {average_score:.2f}") # 계산된 평균을 소수점 둘째 자리까지 출력합니다.
    print(f"최대 점수: {max_score}") # 계산된 최대 점수를 출력합니다.
    print(f"최소 점수: {min_score}") # 계산된 최소 점수를 출력합니다.
    print(f"90점 이상 점수 개수: {count_above_90}개") # 90점 이상인 점수의 개수를 출력합니다.
else:
    print("입력된 점수가 없어 합계와 평균을 계산할 수 없습니다.") # 점수가 없는 경우 메시지 출력
```

done: 번호를 입력받아 pop/remove로 삭제

```python
todo_list = [] # 할 일을 저장할 빈 리스트 초기화

while True: # 프로그램이 종료되기 전까지 무한 반복
    print("\n--- To-Do 리스트 --- ")
    print("1. 할 일 추가")
    print("2. 할 일 목록 보기")
    print("3. 할 일 삭제")
    print("4. 종료")

    try:
        choice = int(input("메뉴를 선택하세요: ")) # 사용자로부터 메뉴 선택을 입력받습니다.

        if choice == 1: # 1번: 할 일 추가
            task = input("추가할 할 일을 입력하세요: ")
            todo_list.append(task) # 리스트의 끝에 새로운 할 일을 추가합니다.
            print(f"'{task}'가 추가되었습니다.")
        elif choice == 2: # 2번: 할 일 목록 보기
            if not todo_list: # 리스트가 비어있는지 확인
                print("등록된 할 일이 없습니다.")
            else:
                print("\n--- 할 일 목록 ---")
                for i, task in enumerate(todo_list): # 리스트의 할 일과 번호를 함께 출력
                    print(f"{i+1}. {task}") # 1부터 시작하는 번호로 출력
        elif choice == 3: # 3번: 할 일 삭제
            if not todo_list:
                print("삭제할 할 일이 없습니다. 먼저 할 일을 추가해주세요.")
            else:
                print("\n--- 할 일 목록 ---")
                for i, task in enumerate(todo_list):
                    print(f"{i+1}. {task}")
                try:
                    task_num_to_delete = int(input("삭제할 할 일의 번호를 입력하세요: "))
                    # 입력받은 번호가 유효한 인덱스 범위 내에 있는지 확인
                    if 1 <= task_num_to_delete <= len(todo_list):
                        deleted_task = todo_list.pop(task_num_to_delete - 1) # 해당 번호의 할 일을 삭제
                        print(f"'{deleted_task}'가 삭제되었습니다.")
                    else:
                        print("유효하지 않은 번호입니다.")
                except ValueError:
                    print("잘못된 입력입니다. 숫자를 입력해주세요.")
        elif choice == 4: # 4번: 종료
            print("프로그램을 종료합니다.")
            break # 무한 루프를 종료하여 프로그램을 마칩니다.
        else: # 1, 2, 3, 4 외의 번호 입력 시
            print("유효하지 않은 메뉴 선택입니다. 1, 2, 3, 4 중 하나를 입력하세요.")
    except ValueError: # 숫자가 아닌 다른 형식의 입력 시
        print("잘못된 입력입니다. 숫자를 입력해주세요.")
```

### 02. 선체 구조 부재 무게 등급 판별   [제어문]

문제: 선박에 사용되는 강판의 무게를 입력받아, 무게에 따라 크레인 운송 등급(A, B, C)을 판별하는 프로그램을 작성하세요.
조건: 50톤 이상은 'A등급', 20톤 이상 50톤 미만은 'B등급', 20톤 미만은 'C등급'으로 분류합니다.

```python
# 사용자로부터 강판의 무게를 입력받습니다.
while True:
    try:
        weight = float(input("강판의 무게를 입력하세요 (톤): "))
        if weight >= 0: # 무게는 0 이상이어야 합니다.
            break
        else:
            print("무게는 0 이상의 값을 입력해야 합니다.")
    except ValueError:
        print("잘못된 입력입니다. 숫자를 입력해주세요.")

# 무게에 따라 크레인 운송 등급을 판별합니다.
if weight >= 50: # 50톤 이상
    grade = 'A등급'
elif weight >= 20: # 20톤 이상 50톤 미만
    grade = 'B등급'
else: # 20톤 미만
    grade = 'C등급'

# 판별된 등급을 출력합니다.
print(f"입력된 강판 무게는 {weight:.2f}톤이며, 운송 등급은 '{grade}'입니다.")
```

### 03.  강재 리스트 반복 출력 및 합계 [반복문]

문제: 리스트에 저장된 여러 개의 강재 무게 데이터를 반복문을 통해 하나씩 출력하고, 전체 강재의 총 무게를 계산하는 프로그램을 작성하세요.

```python
# 사용자로부터 강재 무게 데이터를 입력받습니다.
steel_weights = []
print("강재 무게를 하나씩 입력하세요. 입력을 마치려면 '완료'를 입력하세요.")

while True:
    weight_input = input(f"{len(steel_weights) + 1}번째 강재 무게 (톤): ")
    if weight_input.lower() == '완료':
        break
    try:
        weight = float(weight_input)
        if weight >= 0:
            steel_weights.append(weight)
        else:
            print("무게는 0 이상의 값을 입력해야 합니다.")
    except ValueError:
        print("잘못된 입력입니다. 숫자를 입력하거나 '완료'를 입력해주세요.")

total_weight = 0

if steel_weights:
    print("\n--- 입력된 강재 무게 목록 ---")
    for i, weight in enumerate(steel_weights):
        print(f"{i+1}번째 강재 무게: {weight:.1f} 톤")
        total_weight += weight

    print(f"\n전체 강재의 총 무게: {total_weight:.1f} 톤")
else:
    print("입력된 강재 무게가 없습니다.")
```
